```python
import pandas
import numpy as np
from scipy.optimize import curve_fit
from matplotlib.pyplot import plot,xlabel,ylabel,title,legend

df=pandas.read_csv('HemeBinding3-10A3.csv')#which file you are using, you need to mess with the CSV some to make it readable.
x1=df.Wavelength10 #which wavelength column you are using.  
y1=df.Abs10 #which absorbance column you are using

x1=x1.values.tolist()#numpy likes things in lists
y1=y1.values.tolist()
y1=np.array(y1)
x1=np.array(x1)
```

```python
def expon(x, a, b):
    return a*np.exp(b*-x)
guess=[0,0]
popt, pcov=curve_fit(expon,x1,y1,guess)
print(popt)
```

```python
y2=y1-expon(x1,*popt)
```

```python
plot(x1,y1,label='raw')
plot(x1,expon(x1,*popt), label='fitted')
plot(x1,y2,label='raw-fitted')
legend (loc="upper left")
```
